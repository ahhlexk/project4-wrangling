```python
tweet_id = df_archive['tweet_id']
#df_archive['tweet_id']
#['666020888022790149','666029285002620928','666033412701032449']
#
tweet_errors = {}
import time
start = time.time()
for tweetid in tweet_id: 
    try:
        tweet = api.get_status(tweetid, tweet_mode='extended')
        tweet_json = tweet._json
        with open('tweet_json.txt', 'a') as outfile:  
            json.dump(tweet_json, outfile)
            outfile.write("\n")
        end = time.time()
        total_time = end-start
        print(tweetid, (total_time/60))
    except Exception as e:
        print(str(tweetid) +": "+str(e))
        tweet_errors[tweetid]=tweet_json
        #break
```

```python
input_file = open('tweet_json.txt','r',encoding='utf-8',)
df_list = []
df_errors={}
#retweet_count
#favorite_count
with input_file as f:
    for line in f:
        try:
            data = json.loads(line)
            tweet_id = data['id']
            retweet_count = data['retweet_count']
            favorite_count = data['favorite_count']
            full_text = data['full_text']
            url = data['entities']['media'][0]['url']
            df_list.append({'tweet_id':tweet_id,
                            'retweet_count':retweet_count,
                            'favorite_count':favorite_count,
                            'full_text':full_text,
                            'url':url,
                           })
            df_tweepy = pd.DataFrame(df_list, columns = ['tweet_id',
                                                        'retweet_count',
                                                        'favorite_count',
                                                        'full_text',
                                                        'url'])
        except Exception as e:
            print(str(tweet_id)+': '+str(e))
            df_errors[tweet_id]=line
   
```
